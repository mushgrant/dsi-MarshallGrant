```python
url = 'http://archive.ics.uci.edu.edu/ml/machine-learning-databases/wine-quality/'
```

```python
url
```

```python
dframe_wine = pd.read_csv('datasets/winequality-red.csv',delimiter=';')
```

```python
dframe_wine.head()
```

```python
dframe_wine['alcohol'].mean() #data aggregation that produces a scaler
```

```python
def max_to_min(arr):
    return arr.max() - arr.min()
```

```python
wino = dframe_wine.groupby('quality')
```

```python
wino.describe()
```

```python
wino.agg(max_to_min)
```

```python
max_to_min(wino)
```

```python
wino.agg('mean')
```

# Lecture 45 applying and combining

```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
dframe_wine.sort('alcohol', ascending=False, inplace=True)
```

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker)
```

```python
dframe_wine.head()
```

```python
num_of_qual = dframe_wine['quality'].value_counts()
```

```python
num_of_qual
```

```python
dframe_wine[dframe_wine.alc_content_rank ==1].head(len(num_of_qual))     
```
