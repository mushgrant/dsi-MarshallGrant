# Lecture 49 Kernal density estimation plots

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
```

```python
plt.hist(dataset,alpha=0.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)
x_min = dataset.min() - 2
x_max = dataset.max() + 2

x_axis = np.linspace(x_min,x_max,100)

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**0.2

kernel_list = []

for data_point in dataset:
    
    # Create a kernel for each point and append it to the kernel_list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #scale for plotting
    kernel = kernel /kernel.max()
    kernel = kernel * 0.4 
    
    plt.plot(x_axis, kernel, color ='grey',alpha = 0.5)
    
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list, axis = 0)

fig = plt.plot(x_axis, sum_of_kde,color= 'indianred')

sns.rugplot(dataset)

plt.yticks([])

plt.suptitle("sum of the basis functions")
```

```python
sns.kdeplot(dataset)
```

```python
sns.rugplot(dataset, color = 'black')

for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset,bw=bw,lw=1.8,label=bw)
```

```python
url = 'http://en.wikipedia.org/wiki/Kernel_(statistics)'
```

```python
kernel_options = ['biw','cos','epa','gau','tri','triw']

for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern)

```

# Lecture 51 - Box and Violin Plots

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data = [data1,data2])
```

```python
#normal dist
data_1 = stats.norm(0,5).rvs(100)

#two gamma dist. Concatenated together
data_2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])
#Box plot both data1 and data2
sns.boxplot(data = [data_1,data_2],whis=np.inf)
```

```python
sns.violinplot(data = [data1,data2])
```

```python
sns.violinplot(data = [data_1,data_2])
```

```python
sns.violinplot(data1, inner='stick', )
```

# Lecture 53 Heatmaps and Clusters

```python
flight_dframe = sns.load_dataset('flights')
```

```python
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot('month','year','passengers')
```

```python
flight_dframe
```

```python
sns.heatmap(flight_dframe)
```

```python
sns.heatmap(flight_dframe, annot = True, fmt='d')
```

```python
sns.heatmap(flight_dframe, center = flight_dframe.loc['January',1955])
```

```python
f,(axis1,axis2) = plt.subplots(2,1)
yearly_flights = flight_dframe.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_dframe = pd.concat((years,flights),axis = 1)
year_dframe.columns = ['Year','Flights']

sns.barplot('Year',y='Flights', data=year_dframe, ax=axis1)
sns.heatmap(flight_dframe, cmap='Blues', ax=axis2, cbar_kws={'orientation':'horizontal'})
```

```python
sns.clustermap(flight_dframe)
```

```python
sns.clustermap(flight_dframe, col_cluster=False)
```

```python
sns.clustermap(flight_dframe,standard_scale=1)
```

We standardized the above graph to 1 so that we could really see that summer, regardless of the year, is a busier month to travel. We eliminat the problem of flights increasing every single year.

```python
sns.clustermap(flight_dframe, z_score=1)
```

Above Normalized by Z-score. We subtract the mean and divides by the standard deviatin of each column so that the rows have a mean of zero and a variance of one
