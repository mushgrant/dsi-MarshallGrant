## Navigate to wikipedia and pull down a table of your choice. Here is a sample of what I did to pull down the Harrison Ford movies table to help you out.

```python
import urllib2
from bs4 import BeautifulSoup
import json
import pandas as pd
from pandas import DataFrame 
soup = BeautifulSoup(urllib2.urlopen('https://en.wikipedia.org/wiki/List_of_state_name_etymologies_of_the_United_States').read())
tables = soup.find_all('table', class_="wikitable")
listy = []
for idx, table in enumerate(tables):
   if idx == 0:
       for idx2, row in enumerate(table.find_all('tr')):
           datay = row.find_all('td')
           if len(datay) > 3:
               listy += [{'0':datay[0].text,'1':datay[1].text,'2':datay[2].text,'3':datay[3].text,'4':datay[4].text}]                
stname_org = pd.DataFrame(listy)
stname_org.head(20)
```
