## Step 2: Clean data

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that are too long, and check for missing values (replace them and explain rationale if it makes sense).

```python
# def cleanColumn(x):
#     if x == "artist.inverted":
#         return "artist"
        
#RENAME ARTIST 
x = []             
x = df.columns  
x = list(x)
x[0] = 'artist'
df.columns = x
df
```

```python
#EDIT WEEK COLUMNS TO DISPLAY JUST THE NUMBER THEN, TURN STRING NUMBER TO INTEGER

print(list(df.columns))
x = list(df.columns)
x2 = x[6:]  
new_x = []

for col in x2:
    for letter in col:
            if letter not in ['1','2','3','4','5','6','7','8','9','0']:
                col  =  col.replace(letter,'')
    new_x.append(col)
new_x


finalx = x[:6] + map(int,new_x)

df.columns = finalx
df
```

```python
# create peak time column, time it too to reach highest spot
df['Peak_Time'] = df['date.peaked'] - df['date.entered']
```

```python
# move peak_time to column 6 
cols = df.columns.tolist()
cols =  cols[:6] + cols[-1:] + cols[6:-1] 

df_2 = df[cols]
df_2
```

```python
minSecString = [x.split(":") for x in df_2['time']] 

minSec = [] 
for i in minSecString:
   minSec.append(int(i[0])*60 + int(i[1]))

df_2['seconds'] = pd.Series(minSec)

print(list(df_2.columns)) 
```

```python
cols = df_2.columns.tolist()
cols =  cols[:3] + cols[-1:] + cols[3:-1] 

df_3 = df_2[cols]
df_3
```
