```python
import pandas_datareader.data as web

import datetime
```

```python
prices = web.get_data_yahoo(['CVX','XOM','BP'], 
                               start=datetime.datetime(2014, 1, 1), 
                               end=datetime.datetime(2016, 1, 1))['Adj Close']
```

```python
prices.head()
```

```python
prices.info()
```

```python
prices.plot()
```

```python
prices.CVX.diff(1).autocorr(1)
```

```python
prices.CVX.autocorr(lag=1)
```

```python
prices.CVX.autocorr(lag=50)
```

```python
prices.pct_change().head()
```

```python
from pandas.tools.plotting import autocorrelation_plot
autocorrelation_plot(prices.CVX)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(prices.CVX, lags=10)
```

```python
plot_acf(prices.CVX, lags=50)
```
